**Q1. Describe all three datasets in terms of: number of features, their categories, number of tuples, and missing values.**
**Answer:**

* **Iris Dataset**: Contains 150 tuples with 4 numerical features (sepal & petal measurements) and 1 categorical nominal class label (species). It is a clean dataset with no missing values, widely used for classification.
* **Air Quality Dataset**: Has around 9,000+ tuples with multiple numerical sensor features (CO, NOx, etc.) and temporal features (date/time). It contains many missing values, often represented as -200, requiring preprocessing.
* **Wine Quality Dataset**: Includes 11 numerical features and 1 ordinal categorical output (quality score 0–10). It has about 1,599 (red) and 4,898 (white) samples, with very few or no missing values.

**Q2. In IRIS dataset which non-graphical EDA is performed?**
**Answer:**
Non-graphical EDA in the Iris dataset involves statistical analysis without using plots. We calculate **mean, median, mode, standard deviation, variance, minimum and maximum values** for each feature to understand distribution. A **correlation matrix** is used to identify relationships between features. We also perform **group-wise analysis** by species to compare feature averages across classes. Additionally, checking **data types and class distribution** helps in understanding dataset structure.

**Q3. In Air Quality dataset which graph is used to perform multivariate analysis?**
**Answer:**
For multivariate analysis, a **correlation heatmap** is commonly used in the Air Quality dataset. It visually represents relationships between multiple variables using color gradients, making it easier to detect strong correlations among pollutants. Another useful method is a **scatter plot matrix (pairplot)**, which shows pairwise relationships between variables and helps identify patterns, trends, and outliers across multiple features simultaneously.

**Q4. In Wine Quality dataset can you do non-graphical EDA? Explain your answer.**
**Answer:**
Yes, non-graphical EDA can be effectively performed on the Wine Quality dataset because all features are numerical. We can compute **descriptive statistics** such as mean, median, variance, and quartiles to understand data distribution. **Correlation analysis** helps determine which features influence wine quality the most. We can also group data by quality scores to compare feature averages. Measures like **skewness and kurtosis** further describe the data shape, allowing meaningful insights without using graphs.
