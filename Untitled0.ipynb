## Statistical Testing Overview

### 1. Two-Tailed vs. One-Tailed T-Tests
A **two-tailed test** looks for a difference in either direction, checking if a mean is significantly greater than *or* less than a value. A **one-tailed test** is more specific, testing only if the mean deviates in one pre-defined direction (e.g., strictly "greater than"). You use a two-tailed test when you are unsure of the direction of the effect, whereas one-tailed tests are reserved for when you only care about an increase or a decrease.


### 2. Type I and Type II Errors
A **Type I error** (False Positive) occurs when you reject a true null hypothesis, like telling a healthy person they have a disease. A **Type II error** (False Negative) happens when you fail to reject a false null hypothesis, such as telling a sick person they are healthy. For example, in a trial for a new medicine, Type I means claiming the drug works when it doesn’t, while Type II means missing a drug that actually is effective.

### 3. Power of a Statistical Test
The **power** of a test is the probability that it will correctly reject a false null hypothesis, essentially its ability to detect an effect if one truly exists. It is mathematically defined as $1 - \beta$, where $\beta$ is the probability of making a Type II error. A high power (usually 0.80 or above) means the study is robust enough to find significant results without missing them due to small sample sizes or high noise.

### 4. Characteristics of `stats.ttest_rel`
The `stats.ttest_rel` function performs a paired T-test on two related samples, such as "before and after" measurements on the same group. The **input** requires two arrays of equal length representing the related scores, while the **output** provides a T-statistic and a p-value. The T-statistic indicates the size of the difference relative to variation, and the p-value tells you the probability that this difference occurred by random chance.

### 5. Identifying Power in Python
The `TTestIndPower` or `TTestPower` classes in the `statsmodels.stats.power` module are typically used to calculate statistical power. The output for power is a value between **0 and 1**, where a value closer to 1 signifies a very high likelihood of detecting an effect. Its significance lies in experimental design; it helps researchers determine the minimum sample size needed to ensure their results are reliable and not just due to luck.
