### **1. Need for Hypothesis Testing**

Hypothesis testing is needed to make **statistical decisions** about a population using sample data. It helps determine whether results are due to **chance or actual effects**. It provides a **scientific and objective method** for decision-making. Researchers use it to **test assumptions and validate theories**. It also helps in **comparing groups and analyzing relationships**. Overall, it ensures **accuracy and reliability** in conclusions.

---

### **2. Steps in Hypothesis Testing (with diagram)**

Hypothesis testing starts with stating **H₀ and H₁**, followed by selecting the **significance level (α)**. Then an appropriate test is chosen and data is collected to compute the **test statistic**. Next, the **p-value or critical value** is found. A decision is made to reject or accept H₀. Finally, a conclusion is drawn.

```
Start → State H₀ & H₁ → Choose α → Select Test
      → Compute Statistic → Find p-value
      → Decision (Reject/Accept H₀) → Conclusion
```

### **3. Student t-Distribution**

The Student’s t-distribution is a probability distribution used when the **sample size is small** and population standard deviation is unknown. It was developed by William Sealy Gosset. It is similar to the normal distribution but has **heavier tails**. Its shape depends on **degrees of freedom**. As sample size increases, it approaches the normal distribution. It is widely used in **t-tests**.

---

### **4. Level of Significance & Role of t**

The **level of significance (α)** is the probability of making a **Type I error**, usually 0.05 or 0.01. It defines how strict the test is. The **t-distribution** is used when sample size is small and σ is unknown. It helps calculate the **t-statistic** for testing hypotheses. The t-value is compared with critical values or p-value. This helps decide whether to **reject or accept H₀**.

### **5. Design Hypothesis Test (Age = 19)**

Set H₀: μ = 19 and H₁: μ ≠ 19 (two-tailed test). Choose α = 0.05. Use a **one-sample t-test** since σ is unknown. Collect sample ages and compute the **t-statistic**. Compare it with the critical value or p-value. If p ≤ 0.05, reject H₀; otherwise accept it. Conclude whether average age differs from 19.
