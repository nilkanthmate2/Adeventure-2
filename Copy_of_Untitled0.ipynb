The Chi-square ($\chi^2$) test is a fundamental statistical tool used primarily to determine if there is a significant difference between expected frequencies and observed frequencies in one or more categories.

---

## 1. When to Use the Chi-Square Test
You should opt for a Chi-square test when your data meets these criteria:
* **Categorical Data:** Your variables are nominal (e.g., Eye Color, Department) or ordinal (e.g., Likert scales).
* **Frequency Counts:** You are dealing with raw counts of data points, not means or averages.
* **Independence:** Each observation must be independent of others.
* **Sample Size:** Usually, the "Expected" frequency in any cell should be at least 5 for the results to be reliable.

## 2. Chi-Square Test for Independence
Yes, this is one of its most common uses. It tests whether two categorical variables are related or independent.

**How it works:**
1.  **Formulate Hypotheses:** * $H_0$ (Null): The variables are independent (no relationship).
    * $H_a$ (Alternative): The variables are dependent (a relationship exists).
2.  **Create a Contingency Table:** A grid that cross-tabulates the frequencies of the two variables.
3.  **Calculate Expected Frequencies:** Based on the assumption that $H_0$ is true.
    $$E = \frac{(\text{Row Total} \times \text{Column Total})}{\text{Grand Total}}$$
4.  **Compute the Statistic:**
    $$\chi^2 = \sum \frac{(O - E)^2}{E}$$
    Where $O$ is the Observed frequency and $E$ is the Expected frequency.

## 3. Role of Critical Value and p-value
These values help you decide whether to "Reject" or "Fail to Reject" the Null Hypothesis ($H_0$).

* **Critical Value:** This is a threshold value determined by your **Degrees of Freedom** and your chosen **Significance Level** (usually $\alpha = 0.05$). If your calculated $\chi^2$ is *greater* than the Critical Value, you reject $H_0$.
* **p-value:** This represents the probability of seeing your results (or more extreme) if $H_0$ were true.
    * **$p \le 0.05$:** Statistically significant; reject $H_0$.
    * **$p > 0.05$:** Not statistically significant; fail to reject $H_0$.



## 4. Chi-Square for Correlation?
While the Chi-square test shows **association** (dependence), it does not measure **correlation** in the way a Pearson's $r$ coefficient does.
* **Interpretation:** If the test result is significant ($p < 0.05$), it tells you the two features are not independent—meaning they are "associated."
* **Strength:** To see the *strength* of this association (similar to a correlation coefficient), you would look at metrics like **Cramér's V** or the **Phi Coefficient** alongside the Chi-square result.

## 5. Degrees of Freedom ($df$)
Degrees of freedom represent the number of values in the final calculation that are free to vary.

**Calculation for Independence Test:**
For a table with $r$ rows and $c$ columns:
$$df = (r - 1) \times (c - 1)$$

> **Example:** In a $2 \times 3$ table (e.g., Gender vs. Three Coffee Preferences), $df = (2-1) \times (3-1) = 2$.
